# LOOPS
The process of performing a set of tasks iteratively as per requirement is known as looping.

There are two essential loops in Python.

## While Loop:
while loop is used to execute a block of statements repeatedly subject to a condition being satisfied.
When the expression of the condition is no longer True, the while loop is exited.

```python
## Syntax:
#while condition:
      #code to execute
     #update condition variable
```

```python
count = 0
```

```python
while count < 3:
    print("Data science",count)
    count = count + 1
```

```python
count
```

```python
# Simple while loop
i=1          # counter initialized

while i<6: 
    print('While condition is satisfied for',i) # block of code indented
    i=i+1 # incrementer
    
print("Loop is exited") # outside the while block-of-code
```

```python
i
```

```python
i=1
while i<6:
    print('While condition is true',i)
    print(i)
    i+=1
```

```python
i
```

```python
# While loop with break.
#It is used to stop loop even if while condition is true
i=1
while i<6:
    print(i)
    if i==3:
        break
    i+=1
```

```python
# break – exit loop
num = 0
while num < 10:
    if num == 5:
        break
    print(num)
    num += 1

# continue – skip iteration
num = 0
while num < 5:
    num += 1
    if num == 3:
        continue
    print(num)
```

```python
# While loop with continue statement.
i=0
while i<6:
    i+=1
    if i==3:
        continue
    print(i)
```

```python
# While loop with continue statement.
i=0
while i<8:
    i+=1 # incrementer provided early
    sq = i**2
    if 10<sq<30:
        print("*********Iteration skipped here*******")
        continue
    print(f"Square of {i} = {i**2}")
```

```python
# While loop with else statement.
i=2
while i <6:
    #i+=1
    print(i)
    i=i+1
print("thisss will print no matter the condition")
```

# FOR LOOP
Python for loops are used to loop through an iterable object
(like a list, tuple, set, etc.) and perform the same action for each
item/element.

```python
#SYNTAX
#for item in iterable:
    # code
```

```python
list1 = [1,2,3,4,5]
print(list1)
```

```python
for i in list1:
    print(i)
```

```python
# Looping through a list
fruits=['Apple','Banana','Mango','Watermelon']
for i in fruits:
    print(i)
```

```python
count=[10,12,3,14,25,16,7,8,9,34]
for c in count:
    print(c)
```

```python
# Looping through a string
s='This is data science course'
for i in s:
    print(i, end = ',')
```

# Functions

A function is a reusable block of code that performs a specific task.
Functions make programs:
- Cleaner
- Easy to debug
- Easy to reuse

In Python, a function is defined using the `def` keyword.


## Parameter vs Argument

Parameter: Variable written in the function definition  
Argument: Actual value passed while calling the function


## Three types of functions in Python:-

1) Built-in function :- Python predefined functions that are readily available for use like min() , max() , sum() , print() etc.

2) User-Defined Functions:- Function that we define ourselves to perform a specific task.

3) Anonymous functions : Function that is defined without a name. Anonymous functions are also called as lambda functions. They are not declared with the def keyword.


#### BUILT-IN FUNCTION EXAMPLE

```python
marks = [72, 85, 64, 90, 55]

print("Max Marks:", max(marks))
print("Min Marks:", min(marks))
print("Total Marks:", sum(marks))
```

```python
## Function Returning Multiple Results
def separate_scores(scores):
    passed = []
    failed = []
    
    for s in scores:
        if s >= 40:
            passed.append(s)
        else:
            failed.append(s)
    
    return passed, failed

scores = [32, 55, 68, 22, 90, 41, 37]
separate_scores(scores)
```

## filter() Function

Used to extract elements based on a condition.


```python
temp = [22, 35, 28, 40, 18, 31, 15]

hot_days = filter(lambda t: t > 30, temp)
list(hot_days)
```

```python
seq = [0, 1, 2, 3, 5, 8, 13]  # a list contains both even and odd numbers.
```

```python
# result contains even numbers of the list
result = filter(lambda x: x % 2 == 0,seq)
list(result)
```

```python
# Double all numbers using map and lambda
numbers = (1, 2, 3, 4)
result = map(lambda x: x + x, numbers)
print(list(result))
```

## reduce() Function

Used to reduce a list into a single value.
Imported from functools module.


```python
import functools
from functools import reduce
```

```python
# initializing list
list2 = [1, 3, 5, 6, 2]

# using reduce to compute sum of list
# print("The sum of the list elements is : ", end="")
print(functools.reduce(lambda a, b: a+b, list2))
```

```python
numbers = [1, 2, 3]

result = reduce(lambda a, b: a * b, numbers)
print(result)
```

```python
numbers = [10, 25, 5, 40, 18]

result = reduce(lambda a, b: a if a > b else b, numbers)


print(result)
```

```python
words = ["Python", "is", "awesome","and over"]

result = reduce(lambda a, b: a + " " + b, words)
print(result)
```
